# YOLOv10 — Indoor Dataset Benchmarking

This notebook contains the performance evaluation of **YOLOv10** variants (Nano, Small, Medium) on the Indoor Dataset (7 classes).

## 1. Training Visualization

### 1.1 YOLOv10n Indoor (Nano)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

results_dir = Path("../runs/detect/yolov10n_indoor")
images = ["results.png", "confusion_matrix.png", "BoxPR_curve.png", "BoxF1_curve.png"]

print("Displaying YOLOv10n Indoor Training Results...")
for img_name in images:
    img_path = results_dir / img_name
    if img_path.exists():
        plt.figure(figsize=(10, 8))
        plt.imshow(mpimg.imread(str(img_path)))
        plt.axis("off")
        plt.title(f"YOLOv10n - {img_name}")
        plt.show()
```

## 2. Model Performance Benchmarks

```python
import pandas as pd
import os
from IPython.display import display

csv = "benchmark_yolov10_indoor.csv"
if os.path.exists(csv):
    df = pd.read_csv(csv)
    
    # Styling for a premium look
    fmt = {
        "mAP@0.5": "{:.4f}", 
        "mAP@0.5:0.95": "{:.4f}", 
        "precision": "{:.4f}", 
        "recall": "{:.4f}", 
        "F1": "{:.4f}", 
        "speed_ms/img": "{:.2f}", 
        "size_MB": "{:.1f}"
    }
    
    print("=" * 60)
    print("  YOLOv10 BENCHMARK - Indoor Dataset Comparison (N vs S vs M)")
    print("=" * 60)
    
    styled = (
        df.style.set_caption("YOLOv10 Indoor Dataset Performance Summary")
        .format(fmt)
        .set_properties(**{"text-align": "center", "font-size": "13px"})
        .set_table_styles([
            {"selector": "caption", "props": [("font-size", "16px"), ("font-weight", "bold"), ("padding", "10px")]},
            {"selector": "th", "props": [("background-color", "#1a1a2e"), ("color", "white"), ("padding", "8px")]},
        ])
        .hide(axis="index")
    )
    display(styled)
else:
    print(f"Error: {csv} not found.")
```
